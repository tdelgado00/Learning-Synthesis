```python

def tieBreakByExpansions(path, bests):
    df = pd.read_csv(path, index_col="heuristic")
    print(df[bests])
    
    
def best_generalization_agents_by_budget(path):
    #AGREGAR DESEMPATE POR TRANSICIONE
    rank = solved_by_agent(path)
    best = rank[rank==solved_by_agent(path).max()]
    return best

def solved_by_agent(path):
    df = pd.read_csv(path)
    df = notExceeded(df)
    return df["heuristic"].value_counts(sort=False)

def notExceeded(df):
    return df[df["expansion_budget_exceeded"]==False]
```
